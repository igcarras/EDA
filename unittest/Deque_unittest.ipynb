# Unittest for Dequeue (based on List Python)

To test the functions we will use the __str__ function. For this reason, we have updated the __str__ function to return the same format than Python lists.[texto del enlace](https://)
Moreover, we must assume that the function addLast is correctly implemented because we need a right funtion to load the data to test:


```
class Deque:
  
    def __init__(self):
        self._items=[]

    def __len__(self):
        return len(self._items)

    def __str__(self):
        return str(self._items)

    def isEmpty(self):
        return len(self)==0
 
    def addLast(self,e):
        """adds the element, e, at the end of the dequeue"""
        self._items.append(e)

    def addFirst(self,e):
        self._items.insert(0,e)

    def removeLast(self):
        result=None
        if self.isEmpty():
            print("Error: dequeue is empty!!!")
        else:
            result=self._items.pop()
        return result

    def tail(self):
        """returns the last element of the dequeue"""
        result=None
        if self.isEmpty():
            print("Error: dequeue is empty!!!")
        else:
            #result=self._items[-1]
            result=self._items[len(self)-1]
        return result

    def removeFirst(self):
        result=None
        if self.isEmpty():
            print("Error: dequeue is empty!!!")
        else:
            result=self._items.pop(0)
        return result

    def front(self):
        result=None
        if self.isEmpty():
            print("Error: dequeue is empty!!!")
        else:
            result=self._items[0]
        return result    
```

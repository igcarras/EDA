```
from binarysearchtree import BinarySearchTree, BSTNode

class BSTextend(BinarySearchTree):
    def minimum(self):
        """returns the smallest key of the tree. What is its temporal complexity?"""
        #Complexity: O(log n)
        if self._root is None:
            print('tree is empty')
            return None
        
        node=self._root
        while node.left:
            node=node.left
            
        return node.key     

    def maximum(self):
        """returns the greatest key of the tree. What is its temporal complexity?"""
        #Complexity: O(log n)
        if self._root is None:
            print('tree is empty')
            return None
        
        node=self._root
        while node.right:
            node=node.right
            
        return node.key
    
    def sumKeys(self):
        """returns the sum of all its keys. What is its temporal complexity?"""
        #Complexity: O(n), where n is the size of the tree. 
        #The function has to visit all the nodes of the tree. 
        return self._sumKeys(self._root)
    
    def _sumKeys(self,node):
        if node:
            return node.key + self._sumKeys(node.left) + self._sumKeys(node.right)
        #if node is None
        return 0

    def sumElems(self):
        """returns the sum of all its elements. What is its temporal complexity?"""
        #Complexity: O(n), where n is the size of the tree. 
        #The function has to visit all the nodes of the tree. 
        return self._sumElems(self._root)
    
    def _sumElems(self,node):
        if node:
            return node.elem + self._sumElems(node.left) + self._sumElems(node.right)
        #if node is None
        return 0

    def prints10(self):
        """prints the keys of those nodes whose grandparents' keys are multiply of 10
        What is its temporal complexity""" 
        #Complexity: O(n), where n is the size of the tree. 
        #The function has to visit all the nodes of the tree. 
        self._prints10(self._root)
    
    def _prints10(self,node):
        if node:
            self._prints10(node.left)
            if node.parent and node.parent.parent and node.parent.parent.key%10==0:
                print(node.key, end=' ')
            self._prints10(node.right)
            
    def predecessor(self,key):
        """returns its predecessor node. None if it does not exist"""
        #Complexity: O(log n)
        node=self.find(key)
        if node==None:
            return None
        
        pred=node.left
        if pred:
            while pred.right:
                pred=pred.right
        return pred


    def succesor(self,key):
        """returns its succesor node. None if it does not exist"""
        #Complexity: O(log n)

        node=self.find(key)
        if node==None:
            return None
        
        suc=node.right
        if suc:
            while suc.left:
                suc=suc.left
        return suc

    def _remove(self,node):    
        """Recursive function to remove node.
        Three posible scenarios:
        1) node is a leaf
        2) node only has a child
        3) node has two children"""

        #First case: no children
        if self._numChildren(node)==0:
            #print(node.elem , 'is a leave')
            if node.parent!=None: 
                if node is node.parent.right: #hijo izquierdo del padre:
                    node.parent.right=None
                elif node is node.parent.left: 
                    node.parent.left=None
                node.parent=None
            else:
                self._root=None #tree is empty
                
            return None
        
        #Second case: only one child, the left child
        elif self._numChildren(node)==1:
            grandP=node.parent #grand parent
            #we get the grandchild
            if node.left!=None:
                grandC=node.left   
            elif node.right!=None:
                grandC=node.right
            #link the grandparent to the grandchild    
            if grandP==None:
                self._root=grandC
            elif node is grandP.left:
                grandP.left=grandC
            elif node is grandP.right: 
                grandP.right=grandC
            
            grandC.parent=grandP
            return grandC
            
       
        
        #Third case: two children
        else:
            predecessor=node.left
            while predecessor.right:
                predecessor=predecessor.right
                        
            #we replace the node's elem by the successor's elem
            node.key=predecessor.key
            node.elem=predecessor.elem
            #we remove the succesor from the tree
            return self._remove(predecessor)


    def fe_size(self,key):
        node=self.find(key)
        return self._fe_size(node)

    def _fe_size(self,node):
        if node==None:
            return 0
        else:
            return abs(self._size(node.left)-self._size(node.right))
        

    def fe_height(self,key):
        node=self.find(key)
        return self._fe_height(node)

    def _fe_height(self,node):
        if node==None:
            return 0
        else:
            return abs(self._height(node.left)-self._height(node.right))

    def isSizeBalanced(self):
        return self._isSizeBalanced(self._root)

    def _isSizeBalanced(self,node):
        if node==None:
            return True
        if self._fe_size(node)>1:
            return False
        return self._isSizeBalanced(node.left) and self._isSizeBalanced(node.right)
        
    def isHeightBalanced(self):
        return self._isHeightBalanced(self._root)

    def _isHeightBalanced(self,node):
        if node:
            return self._fe_height(node)<=1 and self._isHeightBalanced(node.left) and self._isHeightBalanced(node.right)
        return True
```

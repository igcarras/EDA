Problema: En la clase MyList, completa la función remove, que toma como argumento un elemento, e, y elimina la primera ocurrencia de este elemento en la lista. La función debe devolver el índice de esta primera ocurrencia. Si el elemento no existe en la lista entonces, la función devuelve -1 y la lista no es modificada. 


```
class SNode:
    def __init__(self,e, next=None):
        self.elem=e
        self.next=next

class MyList:
    def __init__(self):
        self._head=None
        self._tail=None

    def append(self,e):
        newNode=SNode(e)
        if self._head==None:
            self._head=newNode
        else:
            self._tail.next=newNode
        self._tail=newNode

    def __str__(self):
        """Returns a string with the elements of the list"""
        ###This functions returns the same format used 
        ###by the Python lists, i.e, [], ['i'], ['a', 'b', 'c', 'd']
        ###[1], [3, 4, 5]
        nodeIt=self._head
        result='['
        while nodeIt:
            if type(nodeIt.elem)==int:
                result+= str(nodeIt.elem)+ ", "
            else:
                result+= "'"+str(nodeIt.elem)+ "', "
            nodeIt=nodeIt.next
        
        if len(result)>1:
            result=result[:-2]

        result+=']'
        return result


    def remove(self,e):
        """borra la primera ocurrencia de e en la lista y devuelve su índice.
        Si e no existe, devuelve -1.
        
        Mejor caso: Lista vacía, Si la lista no está vacía y e es el primer elemento. O(1)
        Peor caso: e no exista en la lista. O(n)
        
        Complejiad del método remove es lineal"""

        previous=None
        node=self._head
        index=0
        found=False

        while node!=None and found==False:
            if node.elem==e:
                found=True
            else:
                previous=node
                node=node.next
                index+=1

        if node ==None:
            print(e, ' no existe en la lista')
            return -1
        else:
            if node==self._head:  #previous==None
                self._head=self._head.next
            else:    
                #node se el nodo a borrar
                if node==self._tail:
                    self._tail=previous
                
                previous.next=node.next

            return index



    def remove3(self,e):
        #O(n^2)        
        n=len(self)         
        for i in range(n):  
            x=self.getAt(i)     
            if x==e:
                break

        if i in range(n):
            self.removeAt(i)   

        return -1



    def getAt(self,index):
        """Mejor caso: lista vacía, index=0, O(1)
           Peor caso: index=len(self)-1, O(n)
           Complejidad de la función getAt O(n) 
        """
        node=self._head
        i=0
        while node!=None and i<index:
            i+=1
            node=node.next
        
        if node==None:
            print(index , ' no era válido')
            result=None
        else:
            result=node.elem
        
        return result


    def __len__(self):
        node=self._head
        i=0
        while node!=None:
            i+=1
            node=node.next
        
        return i




    def remove2(self,e):
        """borra la primera ocurrencia de e en la lista y devuelve su índice.
        Si e no existe, devuelve -1"""
        #Mejor caso: lista vacía, O(1). Si e es el primer elemento 
        #también se podría considerar como mejor caso. O(1)
        #Peor caso: O(n), no existe el elemento, es necesario recorrer toda la lista.

        #Complejidad de la función: O(n), la complejidad depende del tamaño de
        #la lista. En el peor de los casos es necesario recorrer toda la lista.
        
        index=-1

        found=False

        previous=None
        node=self._head
        #This loop looks for the index of e
        while node and found==False:
            index+=1
            if node.elem==e:
                found=True
            else:
                previous=node
                node=node.next

        if found==False:
            #e does not exist
            index=-1
        else:
            #index is the position of e 
            #node is the node to remove and previous is its previous node
            if previous==None: #node==self._head
                #node is the first node
                self._head=self._head.next
                if self._head==None: #the list only had an element
                    self._tail=None
            else:
                previous.next=node.next
                if node==self._tail:
                    self._tail=previous

        return index





```
